```python
# =====================================================================
# IMPORTAÇÃO DAS BIBLIOTECAS
# =====================================================================
# pandas: Biblioteca essencial para manipulação e análise de dados (tabelas/DataFrames).
import pandas as pd

# train_test_split: Função que divide os dados aleatoriamente em conjuntos de treino e teste.
from sklearn.model_selection import train_test_split

# RandomForestRegressor: O algoritmo de Machine Learning escolhido (Supervisionado/Regressão).
from sklearn.ensemble import RandomForestRegressor

# metrics: Funções para calcular o erro médio (MAE) e a precisão (R²) do modelo.
from sklearn.metrics import r2_score, mean_absolute_error


# =====================================================================
# ETAPA 1: ESCOLHA DA FONTE DE DADOS E OBJETIVO
# =====================================================================
# Os prints abaixo criam um cabeçalho visual centralizado com 60 caracteres de largura.
print("\n" + "="*60)
print(" ETAPA 1: FONTE DE DADOS E OBJETIVO ".center(60, "="))
print("="*60)

# url: Guarda o link direto (formato Raw) do arquivo CSV hospedado no GitHub.
url = "https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/usedcars.csv"

# pd.read_csv: O Pandas baixa o arquivo diretamente da internet e o transforma em uma tabela (DataFrame).
dados = pd.read_csv(url)

# Mostra mensagens informativas no terminal usando f-strings (textos dinâmicos com variáveis).
print(f" -> Dataset 'usedcars.csv' carregado com sucesso direto do GitHub.")
print(f" -> Total de veículos na base: {len(dados)}") # len() conta as linhas da tabela
print(f" -> Colunas originais encontradas:\n    {dados.columns.tolist()}") # Mostra o nome das colunas em formato de lista


# =====================================================================
# ETAPA 2: ANÁLISE DOS DADOS E PRÉ-PROCESSAMENTO
# =====================================================================
print("\n" + "="*60)
print(" ETAPA 2: ANÁLISE E PRÉ-PROCESSAMENTO ".center(60, "="))
print("="*60)

# X: Guarda as "Features" (características do carro) que o modelo usará para aprender.
X = dados[['year', 'model', 'mileage', 'color', 'transmission']]

# y: Guarda o "Target" (alvo), que é o preço do carro. É o que o modelo tentará prever.
y = dados['price']

# pd.get_dummies: Transforma colunas de texto (como 'model', 'color') em números (0 e 1).
# O modelo de machine learning não entende palavras, então o "One-Hot Encoding" cria novas colunas binárias.
# drop_first=True evita redundância de dados (se um carro não é A nem B, ele obrigatoriamente é C).
X = pd.get_dummies(X, columns=['model', 'color', 'transmission'], drop_first=True)

print(f" [OK] Variáveis categóricas convertidas via One-Hot Encoding.")
print(f" [OK] Nova estrutura dos dados (Features): {X.shape[1]} colunas processadas.") # shape[1] diz quantas colunas restaram

# train_test_split: Separa 70% dos dados para o modelo estudar (treino) e guarda 30% escondido (teste).
# random_state=42 garante que a divisão aleatória seja igual toda vez que o código rodar (reprodutibilidade).
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"\n -> Divisão dos dados (70% Treino / 30% Teste):")
print(f"    - Amostras de Treinamento : {X_treino.shape[0]} carros") # shape[0] mostra o número de linhas
print(f"    - Amostras de Teste       : {X_teste.shape[0]} carros")


# =====================================================================
# ETAPA 3: TREINAMENTO DO MODELO
# =====================================================================
print("\n" + "="*60)
print(" ETAPA 3: TREINAMENTO DO MODELO SUPERVISIONADO ".center(60, "="))
print("="*60)
print(" -> Iniciando treinamento com RandomForestRegressor (Sem Redes Neurais)...")

# Cria a instância do modelo. O Random Forest cria várias "árvores de decisão" para chegar no preço final.
modelo_carros = RandomForestRegressor(random_state=42)

# .fit(): É aqui que a mágica acontece. O modelo cruza as características (X_treino) com os preços (y_treino) e aprende os padrões.
modelo_carros.fit(X_treino, y_treino)

print(" [SUCESSO] O modelo de Machine Learning foi treinado com êxito!")


# =====================================================================
# ETAPA 4: VALIDAÇÃO E MÉTRICAS
# =====================================================================
print("\n" + "="*60)
print(" ETAPA 4: VALIDAÇÃO E MÉTRICAS DE DESEMPENHO ".center(60, "="))
print("="*60)

# .predict(): O modelo tenta adivinhar o preço dos 30% de carros que ele NUNCA viu antes (X_teste).
previsoes = modelo_carros.predict(X_teste)

# r2_score: Mede a precisão do modelo. Quanto mais próximo de 100%, melhor o modelo explica a variação dos preços.
r2 = r2_score(y_teste, previsoes)

# mean_absolute_error: Calcula a média do erro. Ex: se der $1200, significa que o modelo erra o preço para mais ou para menos nessa média.
mae = mean_absolute_error(y_teste, previsoes)

# Exibe as métricas na tela formatando os números para terem apenas 2 casas decimais (:.2f).
print(f" -> Coeficiente de Determinação (R²): {r2 * 100:.2f}%")
print(f" -> Erro Médio Absoluto (MAE)       : ${mae:.2f}")

print("\n" + "-"*45)
print(" COMPARAÇÃO PRÁTICA: REAL VS PREVISTO ".center(45, " "))
print("-"*45)

# Cria uma tabela final para o usuário ver visualmente o preço real do carro versus o que o modelo chutou.
comparacao = pd.DataFrame({
    'Preço Real ($)': y_teste,
    'Previsão ($)': previsoes.round(2) # Arredonda os centavos da previsão
})

# .head(5).to_string(index=False): Exibe apenas os 5 primeiros carros e esconde o índice do Pandas para a tabela ficar limpa.
print(comparacao.head(5).to_string(index=False))
print("-"*45 + "\n")
```
